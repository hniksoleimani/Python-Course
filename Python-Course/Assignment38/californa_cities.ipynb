```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("california_cities.csv")
data.head()
```

```python
plt.scatter(data["latd"], 
            data["longd"], 
            s=data["area_total_km2"], 
            alpha=0.5, 
            linewidths=0,
            c=np.log10(data["population_total"]),
            cmap="viridis")
for area in [100, 300, 500]:
    plt.scatter([], [], c='k', alpha=0.3, s=area, label=str(area) + 'km$^2$')
plt.legend(scatterpoints=1, frameon=False, labelspacing=1, title='City Areas')
plt.colorbar(label="log$_{10}$(population)")
plt.clim(3, 7)
plt.xlabel("Latitude")
plt.ylabel("Longitude")
plt.title("Area abd Population of California cities")
plt.savefig("./3_1.jpg")
```
