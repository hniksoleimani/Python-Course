```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import pandas as pd
```

```python


balloon_width = np.random.normal(6, 0.5, 400)  
balloon_length = np.random.normal(6, 0.5, 400)
balloon_weight = np.random.normal(2, 0.5, 400)
```

```python

melon_width = np.random.normal(4, 0.5, 400)  
melon_length = np.random.normal(8, 0.5, 400)
melon_weight = np.random.normal(4, 0.5, 400)
```

```python


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(melon_width, melon_length, melon_weight, color="red", marker="^")
ax.scatter(balloon_width, balloon_length, balloon_weight, color="yellow", label="Bananas")

ax.set_xlabel('Length')
ax.set_ylabel('Width')
ax.set_zlabel('Weight')
plt.savefig("./2_1.jpg")
```
